```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np

dataset=pd.read_csv("train.csv")

dataset.head()
```

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>=1]
features_with_na
```

```python
for features in features_with_na:
    print(features," contain ",np.round(dataset[features].isnull().mean(),4),"% null values")
```

```python
for features in features_with_na:
    data=dataset.copy()
    data[features]=np.where(data[features].isnull(),1,0) #numpy.where(condition[, x, y])
    data.groupby(features)["SalePrice"].median().plot.bar()
    plt.title(features)
    plt.show()
    
```

```python
numerical_features=[features for features in dataset.columns if dataset[features].dtypes!="object"]
print("Number of numerical_features is ",len(numerical_features))
numerical_features
```

```python
dataset[numerical_features].head(5)
```

```python
year_features=[features for features in numerical_features if "Yr" in features or "Year" in features ]
year_features
```

```python
for feature in year_features:
    unique_values = sorted(dataset[feature].unique())
    a = [feature, unique_values]
    print(a)
```

```python
dataset.groupby("YrSold")["SalePrice"].median().plot()
plt.xlabel("Year")
plt.ylabel("Median House Price")
plt.title("Median Price VS Year")
plt.show()
```

```python
for feature in year_features:
    if feature!="Yr_Sold":
        data=dataset.copy()
        # Subtract 'Yr_Sold' from the current feature's values
        #data[feature]=data[feature]-data["YrSold"]
        plt.scatter(data[feature],data["SalePrice"])
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.title(f"{feature}, VS SalePrice")
        plt.show()
```

```python
discrete_features=[features for features in numerical_features if len(dataset[features].unique())<25 and features not in year_features + ["id"]]
print("No of discrete attributes are",format(len(discrete_features)))

discrete_features
```

```python
dataset[discrete_features].head()
```

```python
for features in discrete_features:
    data=dataset.copy()
    data.groupby(features)["SalePrice"].median().plot.bar()
    plt.xlabel(features)
    plt.ylabel("SalePrice")
    plt.title(f"SalePrice VS {features}")
    plt.show()
```

```python
Continous_features=[features for features in numerical_features if features not in discrete_features + year_features +["Id"]]
print(f"No of Continous Attributes are {format(len(Continous_features))} ")

Continous_features
```

```python
dataset[Continous_features].head()
```

```python
for features in Continous_features:
    data=dataset.copy()
    data[features].hist(bins=25)
    plt.title(features)
    plt.xlabel(features)
    plt.ylabel("Count")
    plt.show()
```

## Exploratory Data Analysis 2

```python
for feature in Continous_features:
    data=dataset.copy()
    if 0 in data[feature].unique():
        continue
    elif feature!="SalePrice":
            data[feature]=np.log(data[feature])
            data["SalePrice"]=np.log(data["SalePrice"])
            plt.scatter(data[feature],data["SalePrice"])
            plt.title(f"SalePrice VS {feature}")
            plt.xlabel(feature)
            plt.ylabel("SalePrice")
            plt.show()
    
    
```
